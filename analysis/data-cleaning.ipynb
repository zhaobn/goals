```python
# Import necessary libraries
import pandas as pd
import json
from pprint import pprint

# Read the JSON file
with open('fake-data.json', 'r') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# Function to nicely print state dictionaries
def print_state(state_dict):
    if pd.isna(state_dict):
        print("No state data")
        return
        
    # Parse string if it's a string representation
    if isinstance(state_dict, str):
        state_dict = json.loads(state_dict)
        
    print("Objects state:")
    for obj_id, properties in state_dict.items():
        print(f"\nObject {obj_id}:")
        for prop, value in properties.items():
            print(f"  {prop}: {value}")

# Example: Print state_before for a specific row where it's not null
example_row = df[df['state_before'].notna()].iloc[0]
print("Example state_before:")
print_state(example_row['state_before'])

print("\nExample state_after:")
print_state(example_row['state_after'])
```
